### Припустимо, що на цих даних ми ставимо задачу класифікації, і хочемо класифікувати, до якої групи належить студент (за профілем його відповідей)

```python
mapping_groups = {'ФБ-01, ФБ-02, ФБ-03, ФБ-04, ФБ-05': 0,
  'ФІ-01, ФІ-02': 1,
  'ФІ-03, ФІ-04': 1,
  'ФФ-01, ФФ-03': 2,
  'ФБ-06, ФЕ-01': 2}
reverse_groups = {0: 'ФБ-01, ФБ-02, ФБ-03, ФБ-04, ФБ-05', 1: "ФІ-0X", 2: "інші групи (ФФ, ФЕ)"}
mapping_groups, reverse_groups
```

```python
X = data_encoding
y = data['Група'].apply(lambda x: mapping_groups[x])
```

### Поділимо дані на навчальні та тестові, навчальні потрібні для побудови алгоритму, тестові потрібні для оцінювання його якості

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train, y_train)
```

```python
categorical_columns[5]
```

```python
from sklearn import tree
plt.figure(figsize=(12, 8))
tree.plot_tree(dt)
```

```python
y_predict = dt.predict(X_test)
y_predict
```

```python
y_predict = dt.predict(X_test)
df_predict = pd.DataFrame(zip(y_test, dt.predict(X_test)), columns = ['y_true', 'y_pred'])

df_predict['y_true'] = df_predict['y_true'].apply(lambda x: reverse_groups[x])
df_predict['y_pred'] = df_predict['y_pred'].apply(lambda x: reverse_groups[x])

df_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, dt.predict(X_test)))
```
